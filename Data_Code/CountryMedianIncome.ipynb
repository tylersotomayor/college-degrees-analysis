```python
import requests
import json
import prettytable
import matplotlib.pyplot as plt
import pandas as pd

from api_key import key
url_base = 'https://api.census.gov/data/2017/acs/acs1/spp?'
#25y up educational attainment below HS,above HS,Some College, Bachelor, Graduate
variables = ''
var = ''
variables = [ variables + f'S0201_09{x}E' for x in range(1,6)]
for x in variables:
    var += x
    if x != variables[-1]:
        var += ","
var += ',S0201_235E' #INCOME IN THE PAST 12 MONTHS (IN 2017 INFLATION-ADJUSTED DOLLARS)!!With earnings!!Mean earnings
print(var)
url = f'{url_base}get=NAME,{var}&for=place:*&key={key}' 
print (url)

data = requests.get(url).json()
data = pd.DataFrame(data)
data.columns = data.iloc[0]
data = data[1:-1]
data = data.sort_values('S0201_235E', ascending = False)

var_headers = list(data.columns)[1:-2]
for head in var_headers:
    data[head] = pd.to_numeric(data[head])
data.head()

data.drop(["S0201_235E"],axis =1).set_index('NAME').plot(kind='bar',stacked=True, figsize = (20,5))

plt.ylim(0,115)
plt.legend(('below HS','High School','Some College', 'Bachelor', 'Graduate'),loc = 2, ncol = 5)

plt.title('Education Distribution of Cities by Income per Capita')
plt.ylabel('Percentages (%)')
plt.xlabel('Places')

plt.yticks([0,25,50,75,100])
plt.grid(axis = 'y')

plt.show()
```
